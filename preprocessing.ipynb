```python
import sklearn
import pandas as pd
import requests
import io

file_url = "https://raw.githubusercontent.com/shahabbadihi/IT/master/2-p9vcb5bb.xlsx"
response = requests.get(file_url)

dest = 'local-file.xlsx'

with open(dest, "wb") as filee:
  filee.write(response.content)

df = pd.read_excel(dest)

df.head(100)
```

```python
tdf = df
tdf['cart_id'] = 0
tdf.head(100)
```

```python
userid_list = tdf.user_id.unique().tolist()
print(userid_list)
```

```python
import numpy as np

cartid_start = 1
# number = 0
# product_id 0, user_id 3, cart_id 12

for userid in sorted(userid_list):
    # for idx, row in tdf.iterrows():
    selected_user_rows = tdf[ tdf['user_id'] == userid ]

    # print(len(selected_user_rows))
    len_user_rows = len(selected_user_rows)
    next_user_cartid_start = cartid_start + len_user_rows;

    if (cartid_start == next_user_cartid_start - 1):
        arr_rand_cartids = [cartid_start]
    else:
        np.random.seed(0)
        arr_rand_cartids = np.random.randint(low=cartid_start,
                                         high=next_user_cartid_start - 1,
                                         size=len_user_rows)
    
    tdf.loc[ tdf['user_id'] == userid, 'cart_id'] = arr_rand_cartids
    # if selected_user_rows[3] == user:
    #     row[12] = cart_id
    #     number += 1
    #     if number == 10:
    #         cart_id += 1
    #         number = 0
    # tdf[ tdf['user_id'] == userid ] = selected_user_rows


    cartid_start = next_user_cartid_start
    # number = 0

```
