```python
import pandas as pd
import numpy as np
import requests
import io

file_url = "https://raw.githubusercontent.com/shahabbadihi/IT/master/PreprocessedDataset2.csv"
response = requests.get(file_url)

dest = 'local-file.csv'

with open(dest, "wb") as filee:
  filee.write(response.content)

df = pd.read_csv(dest)
df.head(10)
```

```python
df = df[['user_id', 'product_id', 'product_title', 'cart_id']]
unique_cartids = df.cart_id.unique()
list_carts_df = []

for cartid in unique_cartids:
    list_carts_df.append( df[ df['cart_id'] == cartid ] )
```

```python
dic_carts = {}

for cart in list_carts_df:
    cart_id = cart[['cart_id']].values.tolist()[0][0]
    # print((cart_id))
    # break
    dic_carts[cart_id] = []
    cart2 = cart[['product_id']]
    products = cart2.values.tolist()
    
    for p in products:
       dic_carts[cart_id].append(p[0])
    # print(products)
    # break

print(dic_carts)
```

```python
!pip install apyori
from apyori import apriori
```

```python
list_carts = []
for key, value in dic_carts.items():
    list_carts.append(value)

# print(list_carts)
association_rules = apriori(list_carts, min_length=2, min_support=(2.8/len(list_carts)),
                            min_confidence=0.009, min_lift=1)
association_results = list(association_rules)

print(len(association_results))

```

```python
for i in range(len(association_results)):
  print(association_results[i])
```

```python
rules = []

for i in range(len(association_results)):
    for o in association_results[i].ordered_statistics:
        rules.append(o)

rules
```

```python
def recommendByProductApriori(product_id, rules):
    selected_rules = []
    for r in rules:
        if product_id in r.items_base:
            selected_rules.append(r)

    selected_rules = sorted(selected_rules, key=lambda x: x.confidence, reverse=True)

    recom_products = []
    for r in selected_rules[:2]:
        for i in r.items_add:
            recom_products.append(i)
    
    recom_products = set(recom_products)

    return recom_products


# def isProductInRules(product_id, rules):
#     for r in rules:
#         if product_id in r.items_base:
#             return True
#     return False
```

```python
items_bases = []
for r in rules:
    items_bases.append(r.items_base)

rands = np.random.randint(0, len(items_bases), size=10)
rand_productids = []

for i in rands:
    e = next(iter(items_bases[i]))
    rand_productids.append( e )

# rand_productids = []
# for r in rand_products:
    # if isProductInRules(r, rules):
      # rand_productids.append(r[0][1])

dic_recoms = {}
for r in rand_productids:
    dic_recoms[r] = recommendByProductApriori(r, rules)

print(dic_recoms)
```

```python
likes = 0
count = 0
for productid, recoms in dic_recoms.items():
  print('عنوان کالا:')
  print("\t", df.loc[ df['product_id'] == productid ].product_title.unique()[0])
  print('عناوین کالا های پیشنهادی : ')
  for recommended in recoms:
    print("\t", df.loc[ df['product_id'] == recommended ].product_title.unique()[0])
    count += 1
  print('از چه تعداد از پیشنهادات رضایت داشتید ؟ ')
  likes = likes + int(input())

print('Precision is: ' + str(likes / count))
```
